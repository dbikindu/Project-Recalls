```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import tree
```

```python
pet_finders = pd.read_csv('train.csv')
pet_finders.head()
```

```python
pet_finder= pet_finders.copy()
```

```python
pet_finder.head()
```

```python
pet3_features = pet_finder[['PhotoAmt', 'Age', 'Color2','Color1','Breed1', 'AdoptionSpeed']]
pet3_features.head()
```

```python
# Assign the data to X and y

X = pet3_features[["PhotoAmt", "Age", "Color2", "Color1", "Breed1"]]
y = pet3_features["AdoptionSpeed"].values.reshape(-1, 1)
print(X.shape, y.shape)
```

```python
# Use train_test_split to create training and testing data

### BEGIN SOLUTION
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

### END SOLUTION
```

```python
# Create the model using LinearRegression

### BEGIN SOLUTION
from sklearn.linear_model import LinearRegression
model = LinearRegression()
### END SOLUTION
```

```python
# Fit the model to the training data and calculate the scores for the training and testing data

### BEGIN SOLUTION
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```
